```python
#'Film', 'Genre', 'Rotten Tomatoes Ratings %', 'Audience Ratings %','Budget (million $)', 'Year of release'

sns.set_style("darkgrid")
f, axes = plt.subplots(2,2, figsize=(15,15))
k1 = sns.kdeplot(movies['Budget (million $)'], movies['Audience Ratings %'], ax = axes[0,0])
k2 = sns.kdeplot(movies['Budget (million $)'], movies['Rotten Tomatoes Ratings %'], ax = axes[0,1])
z= sns.violinplot(data= movies[movies.Genre=='Drama'], x='Year of release', y='Rotten Tomatoes Ratings %', ax = axes[1,0])
k4 = sns.kdeplot(movies['Rotten Tomatoes Ratings %'], movies['Audience Ratings %'], cmap='Reds', ax=axes[1,1])

plt.show()
```

```python
sns.set_style("dark", {"axes.facecolor": "black"}) #white, whitegrid, dark, etc.
f, axes = plt.subplots(2,2, figsize=(15,15))

#Plot [0,0]
k1 = sns.kdeplot(movies['Budget (million $)'], movies['Audience Ratings %'], shade=True, shade_lowest=True, cmap='inferno', ax = axes[0,0])
k1b = sns.kdeplot(movies['Budget (million $)'], movies['Audience Ratings %'], cmap='cool', ax=axes[0,0])

#Plot[0,1]
k2 = sns.kdeplot(movies['Budget (million $)'], movies['Rotten Tomatoes Ratings %'], shade=True, shade_lowest=True, cmap='inferno', ax = axes[0,1])
k2b = sns.kdeplot(movies['Budget (million $)'], movies['Rotten Tomatoes Ratings %'], cmap='cool', ax=axes[0,1])

#Plot[1,0]
z= sns.violinplot(data= movies, x='Year of release', y='Budget (million $)', ax = axes[1,0], palette='YlOrRd')

#Plot[1,1]
k4 = sns.kdeplot(movies['Rotten Tomatoes Ratings %'], movies['Audience Ratings %'], shade=True, shade_lowest=False, cmap='Blues_r', ax=axes[1,1])
k4b = sns.kdeplot(movies['Rotten Tomatoes Ratings %'], movies['Audience Ratings %'], cmpa='gist_gray_r', ax=axes[1,1])

plt.show()
```

```python
list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen]['Budget (million $)'])
    mylabels.append(gen)

sns.set_style("whitegrid") #white, whitegrid, dark, etc.    
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)


h = plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.title("Movie Budget Distribution", fontsize=35, color="DarkBlue", fontname="DejaVu Sans")
plt.ylabel("Number of Movies", fontsize=25, color="Pink")
plt.xlabel("Budget of Movies", fontsize=25, color="Olive")
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1, prop={'size':15})
plt.show()
```
